```python
# Importing Statements
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data.info()
```

```python
data.describe()
```

```python
df_num = data[['MarriedID', 'MaritalStatusID', 'GenderID','EmpStatusID', 'DeptID', 'PerfScoreID', 'FromDiversityJobFairID',
       'Salary','EngagementSurvey','EmpSatisfaction','SpecialProjectsCount','Absences','DaysLateLast30']]
print(df_num.info())
plt.figure(figsize=(12,12))
sns.heatmap(data=df_num.corr(), annot=True)
```

```python
df = pd.pivot_table(data= data, values='PerfScoreID', index='ManagerName').sort_values(by='PerfScoreID')
df.plot(kind='barh', figsize=(12, 12))
```

```python
data.Sex.value_counts()/data.Sex.value_counts().sum()
```

```python
data.Sex.value_counts().plot(kind='bar')
```

```python
print(data.RaceDesc.value_counts())
data.RaceDesc.value_counts().plot(kind='bar', figsize=(12,12))
```

```python
ds = data.groupby(data.RecruitmentSource)['GenderID'].value_counts().sort_values()
ds.plot(kind='bar')
```

```python
ds = data.groupby(data.RecruitmentSource)['RaceDesc'].value_counts().sort_values()
ds.plot(kind='bar', figsize=(10,10))
```

```python
df = pd.pivot_table(data=data, values='Salary', columns='Sex')
print(df)
plt.figure(figsize=(12,12))
sns.barplot(data=df)
```

```python
df = pd.pivot_table(data=data, values='Salary', index='Sex', columns='RaceDesc')
print(df)
plt.figure(figsize=(12,12))
sns.barplot(data=df)
```

```python
plt.figure(figsize = (12, 12))
plt.figure(dpi = 150)
plt.title("Relationship between EngagementSurvey and Salary")
sns.regplot(x = "EngagementSurvey", y = "Salary", data = data)
```

```python
plt.figure(figsize = (10, 6), dpi = 150)
plt.title("Relationship between SpecialProjjectsCount and Salary")
sns.regplot(x = "SpecialProjectsCount", y = "Salary", data = data)
```

```python
plt.title("lineplot without using any customization.")
sns.lineplot(x = "EmpSatisfaction", y = "Salary", data = data)
```
