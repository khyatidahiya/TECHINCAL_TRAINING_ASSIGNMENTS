```python
# Replacing Missing values in columns where we have less than 30% missing values
df['Bedroom2'].fillna(df['Bedroom2'].median(),axis=0,inplace=True)
df['Bathroom'].fillna(df['Bathroom'].median(),axis=0,inplace=True)
df['Car'].fillna(df['Car'].median(),axis=0,inplace=True)
df['Landsize'].fillna(df['Landsize'].median(),axis=0,inplace=True)
df['Lattitude'].fillna(df['Lattitude'].median(),axis=0,inplace=True)
df['Longtitude'].fillna(df['Longtitude'].median(),axis=0,inplace=True)
df['Regionname'].fillna(df['Regionname'].mode(),axis=0,inplace=True)
df['CouncilArea'].fillna(df['CouncilArea'].mode(),axis=0,inplace=True)
df['Propertycount'].fillna(df['Propertycount'].median(),axis=0,inplace=True)
```

```python
df['Regionname'].fillna('Southern Metropolitan',inplace=True)
df['CouncilArea'].fillna('Boroondara City Council',inplace=True)
```

```python
# Validaing the Missing value after missing value is treated for few feature columns
print(df.isnull().sum())
```

```python
#Drop Null values from dataframe
dataframe_Housing = df.dropna().sort_values('Date')
```

```python
# Convert the date column to number of days from the date when the house is sold
from datetime import date
days_since_start = [(x-dataframe_Housing['Date'].min()).days for x in dataframe_Housing['Date']]
dataframe_Housing['Days']= days_since_start
```

```python
# Dropping columns which has less correlation to target variable(Price)
df_Housing=dataframe_Housing.drop(['Date','Address','SellerG','Postcode','Landsize','Propertycount'],axis=1)
```

```python
# Convertig Object columns to dummies
df_dummies = pd.get_dummies(df_Housing[['Type','Method','CouncilArea','Regionname']])
```
