```python
# Functions can return multiple values!
def division(num1,num2):
  result = int(num1/num2) # TYPE CASTE
  res = num1%num2
  return result,res # returning 2 independent units

num1,num2 = division(50,2)
print(num1,num2)
```

```python
# Functions can return multiple values!
def divide(a,b):
  q = int(a/b) # TYPE CASTE
  r = a%b
  return (q,r) # returning 1 independent unit-> tuple

a = divide(75,2)
print(a)
print(a[0])
```

```python
# NOT-> POLYMORPHISM
# DOES NOT 
class abc:
  def mul(this, a,b):
    return a*b
  def mul(this, a):
    return a*a

ob = abc()
print(ob.mul(3)) # this should work
print(ob.mul(3,2)) # this should NOT work! PREVIOUS DEF has been OVERRIDDEN by new def!!1
```

```python
class ArithmeticProg:
  def __iter__(self):
    self.base = -10
    self.step = 20
    return self
  def __next__(self):
    self.base = self.base + self.step
    return self.base
```

```python
obj = ArithmeticProg()
i = iter(obj)
```

```python
for j in range(10):
  print(next(i))
```

```python
class Fibo:
  def __iter__(fibo):
    fibo.first = -1
    fibo.second = 1
    return fibo
  def __next__(fibo):
    fibo.third = fibo.first + fibo.second # 0
    fibo.first = fibo.second # 
    fibo.second = fibo.third

    return fibo.third
```

```python
i = iter(Fibo())
for j in range(20):
  print(next(i))
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/cylons/master/data.csv', header=0)
```

```python
diagnosis = data.loc[:,'diagnosis']
encoding_logic = lambda val: 'red' if val=='M' else 'blue'
diagnosis_encoded = diagnosis.map(encoding_logic) # logic will be applied to EVERY element of diagnosis
inputs = data.loc[:,['id','radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean'] ]

plt.figure() # CLEAN the graph buffer before writing anything into it 
sm = pd.plotting.scatter_matrix(inputs, c=diagnosis_encoded, figsize=(25,25))
plt.show()
```

```python
inputs_small = data.loc[:,['radius_mean', 'perimeter_mean','area_mean', 'smoothness_mean', 'symmetry_mean','fractal_dimension_mean'] ]
inputs = inputs.set_index('id')
bins = 12 # relative in nature
plt.figure(figsize=(10,10))
# enumerate([a,b,c]) -> [(0,a),(1,b), (2,c)]
for i, feature in enumerate(inputs.columns):
  rows = 5
  cols = 2
  plt.subplot(rows,cols, i+1 )
  sns.distplot(data[data['diagnosis']=='M'][feature], bins=bins, color='red', label='M')
  sns.distplot(data[data['diagnosis']=='B'][feature], bins=bins, color='blue', label='B')
  plt.legend(loc='upper right')

plt.show()
```
